## LIMPIEZA Y EXPLORACION DE DATOS

```python
#Primero voy a eliminar primero las columnas que siento que estan de mas
vg = videogame.drop(['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Critic_Count', 'User_Count'], axis= 1)
vg
#Borre las ventas de los paises porque ya el dataframe tiene una columna de ventas globales
#y Borre "Critic_Count" y 'User_Count' porque esos solo son la cantidad de criticos que dieron su calificacion
#y considero que con el Critic_Score y User_Score que es el promedio total de calificacion es mas que suficiente
```

```python
#luego eliminare las filas donde las consolas mas irrelevantes esten como el 3DO
#consolas a eliminar debido a que no tienen ni un solo buen juego

borrar = ['NG', 'SCD', 'WS', '3DO', 'TG16', 'GG', 'PCFX']

vg = vg[~vg['Platform'].isin(borrar)]
vg
```

```python
vg['Platform'].value_counts()  #ya no aparacen esas consolas que seleccione para borrar
```

```python
vg.info()
```

```python
#al parecer User_Score tiene la columna en object, a si que la cambiare a float64 igual que columna Critic_Score
vg['User_Score'] = vg['User_Score'].astype('float64')
```

```python
#me sale que no puede convertir el string 'tbd' a float asi que eliminaremos tbd o lo convertiremos en NaN (tbd quiere decir por determinar)

vg[vg['User_Score'] == 'tbd']

#asi que pensandolo mejor eliminare el User_Score ya que generalmente el critic_score es mas acertado
```

```python
vg= vg.drop(['User_Score'], axis= 1) #uso la funcion drop.() dentro coloco la columna a eliminar y le pongo axis =1 para
vg                                    #decirle que sea la columna (axis=0 fila, axis=1 columna)
```

```python
#convertir la columna Year_of_Release a int

vg['Year_of_Release'].astype('int64')
```

```python
#al parecer hay valores NA o inf asi que los encontrare y vere que onda
vg['Year_of_Release'].value_counts()
```

```python
#usare la funcion fillna para llenar las partes NA con cero para aislarlas y ver si les puedo poner el año
vg['Year_of_Release'].fillna(0, inplace = True)
vg['Year_of_Release'] = vg['Year_of_Release'].astype('int64')
```

```python
#despues de analizarlo decidi borrar todos los juegos que les cambie el NA por 0
vg = vg[vg['Year_of_Release'] != 0]
vg['Year_of_Release'].value_counts()
```

```python
vg.head()
```

```python
#despues ordenare el dataframe por nombre del juego
vg.sort_values(by ='Name', inplace= True) #
```

## GRAFICAS


```python
## Cual es la consola que mas juegos tiene
```

```python
vg['Platform'].value_counts() #aqui al usar .value_counts() ya puedo ver que es el ps2 el que tiene mas juegos
                              #seguido del nintendo ds pero quiero verlo en una grafica
```

```python
#quiero ver en una grafica la cantidad de juegos que tiene cada consola
vg_count = vg['Platform'].value_counts()

plt.figure(figsize=(11, 8))   #Tamano de la grafica
plt.bar(vg_count.index, vg_count.values)
plt.xlabel('Consolas')
plt.ylabel('Juegos')
plt.title('Cantidad de Juegos')
plt.xticks(rotation=45) #rotacion de los nombres de las consolas para que se lea mejor

plt.show()
```

```python
#la grafica me muestra de forma visual que el ps2 y el ds son los que mas juegos tienen en su catalogo
```

##Ahora quiero ver que genero es el mas popular

```python
#hare una grafica donde me muestre cuales son los generos mas populares entre todos los juegos

#aqui se calcula la cantidad de cada genero
genre = vg['Genre'].value_counts()

#Ahora se creara la grafica pastel
plt.figure(figsize=(9, 9))  #Tamaño de la gráfica
plt.pie(genre, labels=genre.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  #<-- esta parte del codigo es para que salga el circulo perfecto
plt.title('Generos de videojuegos', y=1.08)
plt.show()
```

```python
#AQUI LA GRAFICA PASTEL NOS MUESTRA QUE EL GENERO DE ACCION ES EL MAS POPULAR O QUE TIENE MAS JUEGOS
#SEGUIDO DE DEPORTES
```

## Ahora quiero saber que años a producido mas juegos

```python
vg['Year_of_Release'].value_counts()#Con .value_counts() nos muestra la cantidad de juegos por año
                                                  
```

```python
#los años que mas juegos an producido son
#2008 con 1427
#2009 con 1426
```

```python
#ahora quiero graficar esas cantidades por años
```

```python
anyo= vg['Year_of_Release'].value_counts().sort_index() #colocamos el codigo en una variable 'anyo'

plt.figure(figsize=(13, 13)) #primero ajustar el tamano de toda la grafica
             #en figsize el 10 es el ancho de 'x' y 9  el ancho de 'y' de toda la grafica (del cuadro blanco)
#crear grafica de barras horizontal
anyo.plot(kind='barh', color='red', width = 0.8)
plt.xlabel("Juegos")
plt.ylabel("AÑO")
plt.title("CANTIDAD DE JUEGOS LANZADOS POR AÑO")
plt.show()
```

```python
#en la grafica podemos visualizar que los años 2008 y 2009 son los que mas juegos se an producido
```
